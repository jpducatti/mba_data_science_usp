### Questão 2 (2 pontos)
Remova do DataFrame todas as linhas cuja soma do total de casos seja menor que 10.000. Imprima quantas linhas foram removidos.

```python
# Reutilizando o dataframe resultante do exericio 1C
# Encontrando as linhas onde os casos acumulados e totais
# são menores que 10.000

# Criando uma mask com as linhas com soma de casos menor 
# que 10.000
mask = (df_drop_prov[df_drop_prov.columns[-1]] < 10000)

# Criando dataframe que não contenha linhas com soma de 
# casos menor que 10.000
df_drop_rows = df_drop_prov[mask == False]

print('O número de linhas removidas é = ',len(df_drop_prov[mask]), 'linhas')
```

### Questão 3 (3 pontos)
Alguns países podem aparecer mais de uma vez no DataFrame, pois podem estar fragmentados em estados ou províncias. Sabendo disso, quais os 5 países que apresentaram o maior número de casos considerando a soma de todos os seus estados e províncias na data mais recente de registro (coluna `3/9/22`)?

```python
# Considerando agora todo o dataframe e removendo
# colunas de latitude, longitude e província. 

df_drop_column  = df.drop(['Province/State','Lat','Long'], axis=1)

# capturando o nome da última coluna
last_column = df_drop_column.columns[-1]

# agrupando os resultados por país, somando 
# os valores do mesmo grupo e ordenando pelo país com maior caso
dfgb_country = df_drop_column.groupby(['Country/Region']).agg(casos = (last_column, 'sum')).sort_values(by = 'casos' , ascending= False)

n = 5 #número de top países com maior casos
print('Os ',n ,' paises com maiores números de casos no registro mais recente são:')
print('\n',dfgb_country.head(n))   

```

### Questão 4 (3 pontos)
Faça o gráfico da evolução do número acumulado de contaminados no Brasil (Brazil) e na Itália (Italy) ao longo do tempo. 

```python
dfgb_country_all_col = df_drop_column.groupby(['Country/Region']).sum()

#capturando as datas da colunas e transformando em array numpy
x = dfgb_country_all_col.columns.to_numpy()

#criando array com dados da italia e brazil
data_italy = (dfgb_country_all_col[dfgb_country_all_col.index == 'Italy']).to_numpy()
data_italy = data_italy.reshape(data_italy.shape[1],)

data_brazil = (dfgb_country_all_col[dfgb_country_all_col.index == 'Brazil']).to_numpy()
data_brazil = data_brazil.reshape(data_brazil.shape[1],)


print('Formato dos dados das datas = ', x.shape)
print('Formato dos dados do Brazil = ',data_brazil.shape)
print('Formato dos dados da Italia = ',data_italy.shape)
```

```python
d = {
    'data' : x,
    'casos_brazil' : data_brazil,
    'casos_italia' : data_italy
}

df_d = pd.DataFrame(d)

df_d['data'] = pd.to_datetime(df_d['data'])
df_d['casos_brazil'] = pd.to_numeric(df_d['casos_brazil'])
df_d['casos_italia'] = pd.to_numeric(df_d['casos_italia'])


print(df_d)
```

```python
from matplotlib import pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import seaborn as sns

fig, ax = plt.subplots(figsize = (15, 10))

#configurando o formato do eixo Y
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f')) 

# plotando os dados de brazil e italia
sns.lineplot(ax = ax, x='data', y='casos_brazil', data=df_d, label='Brazil')
sns.lineplot(ax = ax, x='data', y='casos_italia', data=df_d, label='Italia')

# adicionando titulo ao eixo Y
plt.ylabel('Casos Acumulados de COVID-19', fontsize=18)

#alterando tamanho da fonte dos labels do eixo x, y e legenda
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=20)

plt.show(True)
```
