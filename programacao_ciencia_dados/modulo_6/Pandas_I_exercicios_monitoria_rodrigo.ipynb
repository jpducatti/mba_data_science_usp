### Exercício 8)
Construa um dicionário onde as chaves são os países que já participaram de uma final e os valores são o número de vezes que o país participou da final.

```python
# SOLUÇÃO
resultado_vencedor = DF['Winner'].value_counts()

dic = {}
for chave, valor in resultado_vencedor.items():
  dic[chave] = valor


resultado_vice = DF['RunnersUp'].value_counts()

for chave, valor in resultado_vice.items():
  if chave in dic:
    dic[chave] = dic[chave] + valor
  else:
    dic[chave] = valor

print(dic)
```

### Exercício 9) 
Construa um dicionário com todos os países que já participaram de uma semi-final de copa (os quatro finalistas), onde a chave é o país e o valor corresponde ao número de vezes que o país foi um semi-finalista.

```python
# SOLUÇÃO
resultado_vencedor = DF['Winner'].value_counts()

dic = {}
for chave, valor in resultado_vencedor.items():
  dic[chave] = valor


resultado_vice = DF['RunnersUp'].value_counts()

for chave, valor in resultado_vice.items():
  if chave in dic:
    dic[chave] = dic[chave] + valor
  else:
    dic[chave] = valor

    

resultado_terceiro = DF['Third'].value_counts()

for chave, valor in resultado_terceiro.items():
  if chave in dic:
    dic[chave] = dic[chave] + valor
  else:
    dic[chave] = valor


resultado_quarto = DF['Fourth'].value_counts()

for chave, valor in resultado_quarto.items():
  if chave in dic:
    dic[chave] = dic[chave] + valor
  else:
    dic[chave] = valor

print(dic)
```
